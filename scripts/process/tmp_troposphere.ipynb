```python
import pandas as pd

start_year = 2007
end_year = 2024

base_path = '/root/data/rrr/integrated_weather_dataset/data/processed/Troposphere/'

for year in range(start_year, end_year + 1):
    file_path = f"{base_path}{year}.csv"
    try:
        df = pd.read_csv(file_path)
        
        print(f"Year: {year}")
        
        print("First Row:")
        print(df.head(1))
        
        print("Last Row:")
        print(df.tail(1))
        print("-" * 50)
    except FileNotFoundError:
        print(f"File for year {year} not found.")
    except Exception as e:
        print(f"Error reading file for year {year}: {e}")
```
